### path_embedding.pt

```python
path_embeddings = torch.load(
    "/mnt/drive/hoangbaoan/repos/llm-ids/LMTracker/Phase2/data/model_CUC_20251120175600/path_embedding.pt"
)

print(f"Type of path_embeddings: {type(path_embeddings)}")  # dictionary
print(f"Length of path_embeddings:{len(path_embeddings)}")  # 31283
i = 0
for k, v in path_embeddings.items():
    if i < 1:
        print(f"First item key: {k}")
        print(f"Type of first key: {type(k)}")
        print(f"First item value: {v}")
        i += 1

# Test key accessing
# test_key = torch.tensor([    0, 15479,  2868], device='cuda:0')
# print(path_embeddings[test_key]) # Error because of how tensor are hashes

# Rebuild the dictionary with tuple keys for efficient lookups
path_embeddings_tuple_keys = {tuple(k.cpu().tolist()): tuple(v.cpu().tolist()) for k, v in path_embeddings.items()}

# Now you can query it directly
test_key_as_list = [0, 15425, 10735]
test_key_as_tuple = tuple(test_key_as_list)

print(path_embeddings_tuple_keys[test_key_as_tuple])

pickle.dump(path_embeddings_tuple_keys, open("/mnt/drive/hoangbaoan/repos/llm-ids/LMTracker/Phase2/data/model_CUC_20251120175600/path_embedding.pkl", "wb"), protocol=pickle.HIGHEST_PROTOCOL)
```

```python
model_dir = "/mnt/drive/hoangbaoan/repos/llm-ids/LMTracker/Phase2/data/model_CUC_20251120175600/"


def tensor_to_tuple(file_name):
    torch_file = torch.load(f"{model_dir}{file_name}")
    print(f"Type of path_embeddings of {file_name}: {type(torch_file)}")
    print(f"Length of path_embeddings of {file_name}:{len(torch_file)}")
    print(
        f"First key of {file_name}: {list(torch_file.keys())[0]}. Type: {type(list(torch_file.keys())[0])}"
    )  # tuple
    print(tuple(torch_file[list(torch_file.keys())[0]].cpu().tolist()))  # tensor
    tuplified_file = {k: tuple(v.cpu().tolist()) for k, v in torch_file.items()}
    pickle.dump(
        tuplified_file,
        open(f"{model_dir}{file_name.split('.')[0]}.pkl", "wb"),
        protocol=pickle.HIGHEST_PROTOCOL,
    )


tensor_to_tuple("out_mal_test.pt")
tensor_to_tuple("out_mal_train_val.pt")
```
